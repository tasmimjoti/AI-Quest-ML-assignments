## **Ordinal Encoder**

### **Technique-01**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df4.head()
```

```python
df4.region.unique()
```

```python
area = ["southwest", "southeast", "northwest", "northeast"]
```

```python
ord = OrdinalEncoder(categories = [area])
```

```python
encoded_value = ord.fit_transform(df4[["region"]])
```

```python
new_region = pd.DataFrame(encoded_value, columns = ["Region"])
```

```python
new_region.head()
```

```python
df4.region = new_region
```

```python
df4.head()
```
