## **Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
m = MinMaxScaler()
```

### **Fitting, and then transforming feature step by step**

```python
ms = m.fit(df[["Marketing Spend"]])
ms
```

```python
df["Marketing Spend"] = ms.transform(df[["Marketing Spend"]])
```

```python
df.head()
```

### **Fitting and transforming feature simultaneously**

```python
df["Administration"] = m.fit_transform(df[["Administration"]])
```

```python
df.head()
```

```python
df["Transport"] = m.fit_transform(df[["Transport"]])
```

```python
df.head() # Normalized features
```

## **Robust Scaler**

```python
from sklearn.preprocessing import RobustScaler
```

```python
r = RobustScaler()
```

```python
t1 = r.fit(df2[["Administration"]])
t1
```

```python
df2["Marketing Spend"] = r.fit_transform(df2[["Marketing Spend"]])
df2["Administration"] = r.fit_transform(df2[["Administration"]])
df2["Transport"] = r.fit_transform(df2[["Transport"]])
```
